### PROBLEM STATEMENT

Fictitious Pharmaceutical Company ‘Ollopa Pharmaceuticals’ manufactures the Active Pharmaceutical Ingredient (API) for its small molecule blockbuster drug ‘Asclepius’ by stepwise organic synthesis in a processing plant. The Manufacturing Science Group have been asked to optimise the production process to increase the yield of the active ingredient by monitoring the reaction endpoint using process analytical technology. The effect of changing process conditions and their relationship with reaction kinetics need to be monitored to achieve this. The Manufacturing Science Group have identified 4 Process Variables which they would like to investigate to see if there is any correlation between them and the reaction endpoint. The importance of identifying the reaction endpoint is critical for reactions where over reaction may lead to impurities being formed and the underreaction may lead to leftover starting materials whcih may not be purged from the process. 

### 1.CREATE THE VARIABLES

**Critical Process Parameters**

|RAP(%)     |Reaction Temp      |  pH                   | CO2                     |
|:-------:  |:-----------------:|:--------------------: |:-----------------------:|
| 0.2-99.9% |          0-25     | pH = 4.5-7.5          | 0-0.5%                  |

#### **Variables Explained**

**RAP** *Relative area %*: The % of Active ingredient produced during the course of the Reaction relative to the starting material (as the level of active increases, the level of strting matrial decreases. *We hope to determine the effects of the following variables on its rate*

**Reaction Temp:** Temp the reaction reaches (it is an exothermic reaction, i.e. one in which heat is produced)
**Pre crystallisation pH:** The pH of the batch during the reaction
**CO2(%):** CO2 Levels produced during the reaction as a by-product

The In-Process variables are the parameters which influence the yield and quality of the active ingredient. A manufacturing process which deviates from these parameters may have low yields and/or high impurity levels.

For the purposes of creating a dataset we will investigate 100 data points of variables i.e. we will monitor a reaction over 10hrs, with the process variables being measured ten times per hour using inline process analytical techniques.

The effect or contribution (if any) of these variables on the reaction will be investigated.




Import of python packages i.e. Numpy, Pandas, Scipy, odeint (a sub package of scipy for differential equations), Seaborn. The number of observations was set at 100 and a random seed generated. It was necessary to set a random seed in order to preserve the consistency of the data - since random variables were created each time the Jupyter Notebook was ran.

The Relative Area Percent variable was created by trial and error using the ```odeint``` function. dy/dt vs time could then be plotted

pH and CO2 Variables were created using a random normal function rounded to 2 decimal places




````python
import numpy as np
import pandas as pd
import scipy as sp
from scipy.integrate import odeint #odeint is a library which allows us to create and solve differential equations in python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline  

#number of observations

size = 100 

#Random seed set to maintain repeatability of the dataset
np.random.seed(876556788)

#List of variables
Time = np.arange(1,101) # Number of data points from start of reaction

#RAP Will be calculated as a model to mimic a first order kinetic curve

#RxnTemp data will be created to correlate with the change in RAP over time.

pH = np.round(np.random.normal(loc=7.0, scale=0.25, size=size),2)

CO2 = np.round(np.random.normal(loc=0.25, scale=0.025, size=size),2)

````

**We will now generate a function to create a differential curve to model the increase in RAP during the reaction**

The 'model' is a function which returns values of dy/dt for y and t. Initial value fro y was set at 0 and ```np.linspace``` returns evenly spaced numbers over an interval. 'odeint' then calculates y(t). The resulting variables are then plotted as y V t defining y as RAP for the purposes of our dataset.

````python
#Credit: Adapted from (https://apmonitor.com/pdc/index.php/Main/SolveDifferentialEquations)
#Placeholder for some simple code for graphing a differential curve to be adapted
#in order to mimic first order reaction  kinetics

# function that returns dy/dt
def model(y, t):
    dydt = -y + 100.0
    return dydt

# initial condition for y
y0 = 0

# time points
t = np.linspace(1, 10, 100) 

# solve ODE
y = (odeint(model,y0,t)) #uses odeint function to model the relationship between dy/dt and t

# plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('RAP(%)')
plt.show()
````

````python
#Output of y as a multidimensional array.
y
````

'y' Will need to be flattened into a single dimension array in order to be concatenated with the other variables using ```np.flatten()```

````python
RAP = y.flatten() #Converts multidimensional array to single dimensional array, allowing it be be concatenated with the 
# other random arrays in our dataset
RAP
````

Another dy/dt v t model was created in order to model a change in reaction temperature over time. A lot of experimentation was used to find the right function for dy/dt. Eventually the ```np.cos```[](https://docs.scipy.org/doc/numpy/reference/generated/numpy.cos.html) command proved useful: 

Since it is a simulated dataset, the objective is to create at least two variables which are correlated and RAP and Temperature were the two variables chosen, hence the need for them to be simulated using differential equations

````python
#Some experimentation to try to model Temp data to fit with RAP data
#Lots of random experimentation with help from (https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.math.html)

def model(z, t):
    dydt = (np.cos(1/t) - 1)*(-40)
    return dydt

# initial condition for y
z0 = 0

# time points
t = np.linspace(1, 10, 100) 

# solve ODE
z = (odeint(model,z0,t)) #uses odeint function to model the relationship between dy/dt and t

#RAP = pd.DataFrame(y)
# plot results
plt.plot(t,z)
plt.xlabel('time')
plt.ylabel('RxnTemp')
plt.show()
````

````python
RxnTemp = z.flatten() #Converts mulitdimensional array to single dimensional array, allowing it be be concatenated with the 
# other random arrays in our dataset
RxnTemp
````

All the variables were output as NP arrays for inspection together

````python
#Output the Data as NUMPY arrays
print(('Time(hrs)=', Time), ('RAP =', RAP), ('pH =', pH), ('Reaction Temp(C) =', RxnTemp), ('CO2 =', CO2))
````

### 2. CREATE A DATAFRAME OF ALL VARIABLES

**Each of the 4 numpy arrays will now be concatenated to create 1 dataframe**

[1][Column names](https://docs.python.org/3.4/tutorial/datastructures.html#dictionaries)

All of variables were concatenated to form one dataframe. Easily done using Pandas dataframe command.Names added to each column also. The purpose of this was to assemble a dataframe of simulated variables representing the hypothetical data gathered for the reaction of one batch of active ingredient.

**The following function ```sns.pairplot``` using the Seaborn package, creates a scatterplot matrix of variables and includes a density estimate (in the diagonal axis). It's a useful package to visualise data with multiple variables and get a quick idea of the relationships between variables. [11]**

The Seaborn pairplot function enables a broad overview of how each variable relates to one another, showing scatter plots, linear relationships and Kernel density estimates. Kernel density estimates provide an estimate of how the data in each variable is distributed whcih may be useful to ascertain the types of subsequent swtatistical tests to be performed on the data

**Example scatter plot of pH V RAP**

````python
g = sns.FacetGrid(Batchdf, height=5, )
g.map(plt.scatter, "pH", "RAP", s=50, linewidth=.5, edgecolor="white")
g.add_legend();

````

**Plot The linear regression models for each variable against yield in order to determine if any linear relationships exist between variables**

Using the Seaborn package to plot variables and output the regression equation. Of particualr value is the R^2 value which tells us if there is correlation between variables 

````python
# get coefficients of linear fit

# Linear regression for pH V RAP
slope, intercept, r_value, p_value, std_err = stats.linregress(Batchdf['pH'],Batchdf['RAP'])

# Set line label for legend
ax = sns.lmplot(x='pH', y='RAP', data=Batchdf)

plt.title("pH V RAP {}: y={:.1f}x+{:.1f} (p: {:.1f}, R^2: {:.2f})".format(ax, slope, intercept,  p_value, r_value))
plt.show() # use of Seaborn to plot the relationships between data and generate the regression equation
````

````python
# get coefficients of linear fit

# Linear regression for Reaction Temp V RAP
slope, intercept, r_value, p_value, std_err = stats.linregress(Batchdf['Reaction Temp(C)'],Batchdf['RAP'])

# use line_kws to set line label for legend
ax = sns.lmplot(x='Reaction Temp(C)', y='RAP', data=Batchdf)

plt.title("RAP V Reaction Temp(C) {}: y={:.1f}x+{:.1f} (p: {:.1f}, R^2: {:.2f})".format(ax, slope, intercept,  p_value, r_value))
plt.show()
````

````python
# get coefficients of linear fit

# Linear regression for IPC Result(%) V RAP
slope, intercept, r_value, p_value, std_err = stats.linregress(Batchdf['CO2'],Batchdf['RAP'])

# use line_kws to set line label for legend
ax = sns.lmplot(x='CO2', y='RAP', data=Batchdf)

plt.title("RAP V CO2 {}: y={:.1f}x+{:.1f} (p: {:.1f}, R^2: {:.2f})".format(ax, slope, intercept,  p_value, r_value))
plt.show()
````

## DISCUSSION OF LINEAR REGRESSION

It appears that the linear regression models indicate that the only relationship between variables which is linear is RAP V Reaction temp with a R^2 Value of 0.98, i.e. as the reaction progresses the temp increases. Display of the Correlation coefficients for each variable and it's relationship with RAP will help us determine the next steps in the analysis of the data.

Uses the ```sp.stats.pearsonr()``` function

````python
 #Calculates the Pearson correlation coefficient for all variables V RAP and the *p* value for non-correlation
pHVRAP = sp.stats.pearsonr(pH,RAP)
RxnTempVRAP = sp.stats.pearsonr(RxnTemp,RAP)
CO2VRAP = sp.stats.pearsonr(CO2,RAP)

print ('pH V RAP: Correlation Coefficient', pHVRAP)
print ('Reaction Temperature V RAP: Correlation Coefficient', RxnTempVRAP,)
print ('CO2 V RAP: Correlation Coefficient', CO2VRAP)

````

The correlation coefficient for Reaction Temperature V RAP, calculated here using the 'scipy.stats.pearson' function shows a positive correlation between the two variables. This combined with the low *p* value confirms the hypothesis that the two variables are correlated.

## CONCLUSION

The relationship between temperature and RAP shows us that as the reaction progresses the temperature rises reaching a maximum as the reaction is complete. Knowing how as the reaction progresses, the temperature increases and levels off when complete will enable us to use the temperature as a guide when the reaction is complete and therefore when to sample for analysis to be confirmed offline in the laboratory.

Such techniques can predict and determine the cause of out of control processes in many industries. The Hotellings T^2 statistic (the multivariate version of the students t-test statistic) and the Q statistic (test for randomness) are two methods which are used for the identification of out of control processes. [10] These two statistics are determined on data on which principal component analysis has been carried out. Principal Component analysis is a data analytic technique as described by Jackson, 1991, where correlated variables are linearly transformed into 'principal components' which themselves are not correlated to eachother. PCA is in effect a technique for differentiating variables from eachother and in the process *reducing dimensions*. While there are only 4 variables in our data set and two which are correlated, our first principal component would be RAP/Temp as both are correlated variables. 

PCA may be carried out using the ```sklearn.decomposition``` package and ```PCA``` class in Scikitlearn.
